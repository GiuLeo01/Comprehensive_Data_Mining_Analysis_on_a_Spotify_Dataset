```python
import numpy as np
import keras as k
import tensorflow as tf
import matplotlib.pyplot as plt
from keras import layers as l
from augmentation import *
from utils import *
```

```python
np.load('nn_models/cnn_ts_128_FINAL/logs.npz')
```

```python
train_acc = np.load('nn_models/cnn_ts_128_FINAL/logs.npz')['tot_acc']
train_loss = np.load('nn_models/cnn_ts_128_FINAL/logs.npz')['tot_train_loss']
val_acc = np.load('nn_models/cnn_ts_128_FINAL/logs.npz')['tot_val_acc']
val_loss = np.load('nn_models/cnn_ts_128_FINAL/logs.npz')['tot_val_loss']
```

```python
epochs = range(1, len(val_acc)+1)

plt.figure(figsize=(12,5))
plt.plot(epochs, train_loss, label='Training Loss', c='red')
plt.plot(epochs, val_loss, label='Validation Loss', c='blue')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(fontsize='large')
plt.axvline(x=np.argmax(val_acc)+1, c='black', linestyle='--', linewidth=2)
plt.text(x=np.argmax(val_acc)-800, y=8, s=f'Stopped at Epoch {np.argmax(val_acc)+1}', fontsize='x-large')
plt.text(x=np.argmax(val_acc)-800, y=7.2, s=f'Validation Loss: {round(val_loss[np.argmax(val_acc)], 2)}', c='blue', fontsize='x-large')
plt.text(x=np.argmax(val_acc)-800, y=6.4, s=f'Training Loss: {round(train_loss[np.argmax(val_acc)], 2)}', c='red', fontsize='x-large')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(epochs, train_acc, label='Training Accuracy', c='red')
plt.plot(epochs, val_acc, label='Validation Accuracy', c='blue')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(fontsize='large')
plt.axvline(x=np.argmax(val_acc)+1, c='black', linestyle='--', linewidth=2)
plt.text(x=np.argmax(val_acc)-800, y=0.25, s=f'Stopped at Epoch {np.argmax(val_acc)+1}', fontsize='x-large')
plt.text(x=np.argmax(val_acc)-800, y=0.225, s=f'Validation Accuracy: {round(val_acc[np.argmax(val_acc)], 2)}', c='blue', fontsize='x-large')
plt.text(x=np.argmax(val_acc)-800, y=0.20, s=f'Training Accuracy: {round(train_acc[np.argmax(val_acc)], 2)}', c='red', fontsize='x-large')
plt.show()
```
