```python
unbalanced_dataset = dataset.copy(deep=True)
unbalanced_dataset['marker'] = 0
for i in range(unbalanced_dataset.shape[0]):
    if unbalanced_dataset.iloc[i]['explicit'] == True:
        unbalanced_dataset.at[i,'marker'] = np.random.binomial(size=1,n=1,p=0)[0]
```

```python
np.unique(unbalanced_dataset[unbalanced_dataset['marker'] == 1]['explicit'])
```

```python
unbalanced_dataset_dropped = unbalanced_dataset[unbalanced_dataset['marker'] == 0].drop('marker',axis=1)
```

```python
unbalanced_dataset_dropped = unbalanced_dataset_dropped.set_index(np.arange(unbalanced_dataset_dropped.shape[0]))
```

```python
unbalanced_dataset_dropped.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()


unbalanced_dataset_dropped['enc_genre'] = encoder.fit_transform(unbalanced_dataset_dropped['genre'])
encoding_dict = {i:class_ for class_ ,i in zip(encoder.classes_, range(141))}
```

```python
unbalanced_dataset_dropped
```

```python
encoding_dict
```

```python
X = unbalanced_dataset_dropped[['duration_ms', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
                                'liveness', 'valence', 'tempo', 'time_signature', 'tempo_confidence', 'time_signature_confidence', 'key_confidence', 'mode_confidence', 'enc_genre']]
y = unbalanced_dataset_dropped['explicit']

X_cont = unbalanced_dataset_dropped[['duration_ms', 'popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness',
                                'liveness', 'valence', 'tempo', 'tempo_confidence', 'time_signature_confidence', 'key_confidence', 'mode_confidence']]
```

```python
X.shape, y.shape
```

```python
X = X.astype({'duration_ms': 'float64', 'popularity': 'float64', 'enc_genre':'int64'})
```

```python
X.select_dtypes('int64')
```

```python
X_cat = X_cont.copy(deep=True)

numeric_columns=list(X.select_dtypes('float64').columns)
categorical_columns=list(X.select_dtypes('int64').columns)


enc_genre = pd.get_dummies(X['enc_genre'])
enc_genre.columns = [encoding_dict[i] for i in range(114)]
X_cat[enc_genre.columns] = enc_genre

enc_key = pd.get_dummies(X['key'])
enc_key.columns = [f'key_{n}' for n in enc_key.columns]
X_cat[enc_key.columns] = enc_key

enc_ts = pd.get_dummies(X['time_signature'])
enc_ts.columns = [f'ts_{n}' for n in enc_ts.columns]
X_cat[enc_ts.columns] = enc_ts

enc_mode = pd.get_dummies(X['mode'])
enc_mode.columns = [f'mode_{n}' for n in enc_mode.columns]
X_cat[enc_mode.columns] = enc_mode

```

```python
X_cat.replace({False: 0, True: 1}, inplace=True)
X_cat.info()

y.replace({False: 0, True: 1}, inplace=True)
```

```python
X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(X_cat, y, stratify=y, random_state=8, test_size=0.25)

X_train_cont, X_test_cont, y_train_cont, y_test_cont = train_test_split(X_cont, y, stratify=y, random_state=8, test_size=0.25)
```

```python
pd.DataFrame(X_test_cat).shape, pd.DataFrame(X_test_cont).shape
```

```python
X_test_cat.select_dtypes('int64').shape
```

```python
from sklearn.preprocessing import StandardScaler

sca = StandardScaler()
X_train_cont_sca = sca.fit_transform(X_train_cont)
X_test_cont_sca = sca.transform(X_test_cont)

add_train = pd.DataFrame(X_train_cat.select_dtypes('int64')).set_index(np.arange(X_train_cat.shape[0]))
add_test = pd.DataFrame(X_test_cat.select_dtypes('int64')).set_index(np.arange(X_test_cat.shape[0]))

X_train_cat_sca = pd.concat([pd.DataFrame(X_train_cont_sca).copy(deep=True), add_train], axis=1)
X_test_cat_sca = pd.concat([pd.DataFrame(X_test_cont_sca).copy(deep=True), add_test], axis=1)

X_train_cat_sca = X_train_cat_sca.drop(['popularity', 'duration_ms'], axis=1)
X_test_cat_sca = X_test_cat_sca.drop(['popularity', 'duration_ms'], axis=1)

X_train_cat_sca.columns = X_train_cat.columns
X_test_cat_sca.columns = X_test_cat.columns
```

```python
X_train_cont_sca.shape
```

```python
X_train_cont_sca = pd.DataFrame(X_train_cont_sca)
X_test_cont_sca = pd.DataFrame(X_test_cont_sca)
X_train_cont_sca.columns = X_train_cont.columns
X_test_cont_sca.columns = X_test_cont.columns
```

```python
list(X_test_cont_sca.columns)
```

```python
X_train_cont_sca = X_train_cont_sca.drop(['time_signature_confidence', 'key_confidence', 'mode_confidence'], axis=1)
X_test_cont_sca = X_test_cont_sca.drop(['time_signature_confidence', 'key_confidence', 'mode_confidence'], axis=1)
```
