```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df_raw = pd.read_csv('../dataset_tab/merged_dataset.csv')
```

```python
df_raw.columns
```

```python
df_raw
```

```python
y = df_raw['popularity_y']
X_raw = df_raw.drop('popularity_y', axis=1)
```

```python
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X_raw, y, test_size=0.3, random_state=100, stratify=df_raw['genre'])
```

```python
X_train_raw.columns
```

```python
X_train_cont = X_train_raw[['duration_ms', 'danceability', 'energy', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'tempo_confidence', 'time_signature_confidence',
       'key_confidence', 'mode_confidence', 'popularity_x', 'followers']]

X_test_cont = X_test_raw[['duration_ms', 'danceability', 'energy', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'tempo_confidence', 'time_signature_confidence',
       'key_confidence', 'mode_confidence', 'popularity_x', 'followers']]
```

```python
X_train_cat = X_train_raw[['explicit', 'key', 'mode', 'time_signature', 'genre']]
X_test_cat = X_test_raw[['explicit', 'key', 'mode', 'time_signature', 'genre']]
```

## Senza One hot encoding

```python
X_train_cat = X_train_cat.astype({'explicit': 'category', 'key': 'category', 'mode':'category', 'time_signature':'category', 'genre':'category'})
X_test_cat = X_test_cat.astype({'explicit': 'category', 'key': 'category', 'mode':'category', 'time_signature':'category', 'genre':'category'})
```

```python
scaler = StandardScaler()
X_train_cont_norm = scaler.fit_transform(X_train_cont)
X_test_cont_norm = scaler.transform(X_test_cont)

# esegui fino a qui pe rone hot encoding
```

```python
X_train = pd.DataFrame(np.concatenate([X_train_cont_norm, X_train_cat], axis=1))
X_test =  pd.DataFrame(np.concatenate([X_test_cont_norm, X_test_cat], axis=1))
```

```python
X_train.columns = ['duration_ms', 'danceability', 'energy', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'tempo_confidence', 'time_signature_confidence',
       'key_confidence', 'mode_confidence', 'popularity_x', 'followers'] + ['explicit', 'key', 'mode', 'time_signature', 'genre']
X_test.columns = ['duration_ms', 'danceability', 'energy', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'tempo_confidence', 'time_signature_confidence',
       'key_confidence', 'mode_confidence', 'popularity_x', 'followers'] + ['explicit', 'key', 'mode', 'time_signature', 'genre']
```

```python
X_train = X_train.astype({'explicit': 'category', 'key': 'category', 'mode':'category', 'time_signature':'category', 'genre':'category',
                          'duration_ms' : 'float32', 'danceability': 'float32', 'energy': 'float32', 'loudness': 'float32', 'speechiness': 'float32',
       'acousticness': 'float32', 'instrumentalness': 'float32', 'liveness': 'float32', 'valence': 'float32', 'tempo': 'float32', 'tempo_confidence': 'float32', 'time_signature_confidence': 'float32',
       'key_confidence': 'float32', 'mode_confidence': 'float32', 'popularity_x': 'float32', 'followers': 'float32'})

X_test = X_test.astype({'explicit': 'category', 'key': 'category', 'mode':'category', 'time_signature':'category', 'genre':'category',
                          'duration_ms' : 'float32', 'danceability': 'float32', 'energy': 'float32', 'loudness': 'float32', 'speechiness': 'float32',
       'acousticness': 'float32', 'instrumentalness': 'float32', 'liveness': 'float32', 'valence': 'float32', 'tempo': 'float32', 'tempo_confidence': 'float32', 'time_signature_confidence': 'float32',
       'key_confidence': 'float32', 'mode_confidence': 'float32', 'popularity_x': 'float32', 'followers': 'float32'})
```

```python
# salvo i dati
X_train.to_csv('data_no_onehot/X_train', index=False)
X_test.to_csv('data_no_onehot/X_test', index=False)
y_train.to_csv('data_no_onehot/y_train', index=False)
y_test.to_csv('data_no_onehot/y_test', index=False)
```
